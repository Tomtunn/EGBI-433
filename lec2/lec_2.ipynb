```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# plot histogram
def contrast_stretching(img, low, high):
    img_min = np.min(img)
    img_max = np.max(img)
    new_img = (img - img_min) * ((high - low) / (img_max - img_min)) + low
    new_img = new_img.astype(np.uint8)
    return new_img
```

```python
# import image
img = cv2.imread('penguin.png')
# plot histogram
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].imshow(img)
axs[0].set_title('Original Image')
axs[1].hist(img.ravel(), 256, [0, 256])
axs[1].set_title('Histogram')
plt.show()
```

```python
contrast_img = contrast_stretching(img, 0, 255)

# plot histogram
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].imshow(contrast_img)
axs[0].set_title('Contrast Image')
axs[1].hist(contrast_img.ravel(), 256, [0, 256])
axs[1].set_title('Histogram')
plt.show()
```
