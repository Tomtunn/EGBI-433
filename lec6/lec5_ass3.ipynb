```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def unsharpe_mask_filtered(img, ksize, sigma, alpha):
    # blur
    blur = cv2.GaussianBlur(img, ksize, sigma)
    # unsharpe mask
    mask = img - blur
    # sharpened image
    sharpened = img + (alpha * mask)
    return sharpened
    # return blur
```

```python
ksize = (5, 5)
sigma = 1

img = cv2.imread('penguin2.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

sharped_25 = unsharpe_mask_filtered(img, ksize, sigma, alpha=2.5).astype(np.uint8)
sharped_100 = unsharpe_mask_filtered(img, ksize, sigma, alpha=10.00).astype(np.uint8)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 15))
axs[0].imshow(img, cmap='gray')
axs[0].set_title('original')
axs[1].imshow(sharped_25, cmap='gray')
axs[1].set_title('alpha=2.5')
axs[2].imshow(sharped_100, cmap='gray')
axs[2].set_title('alpha=10.0')
plt.show()
```

# Dicussion
the unmasked technique is use in edge sharpening, the output are depend on the alpha value, the higher the alpha value, the more sharpening the image will be, however, the higher the alpha value, the more noise will be added to the image.
